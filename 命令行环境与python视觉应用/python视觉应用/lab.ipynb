```python
from PIL import Image,ImageEnhance
from scipy import ndimage
from skimage import transform
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_ori=Image.open('0.jpg')
img_enhanced=ImageEnhance.Contrast(img_ori).enhance(3.0)
plt.imshow(np.hstack((img_ori,img_enhanced)))
plt.title('')
plt.axis('off')
plt.show()
```

```python
np_img=np.array(Image.open('1.jpg'))
r=np_img.copy()
r[:,:,(1,2)]=0
g=np_img.copy()
g[:,:,(0,2)]=0
b=np_img.copy()
b[:,:,(0,1)]=0
plt.imshow(np.hstack((np_img,r,g,b)))
plt.title('RGB')
plt.axis('off')
plt.show()
```

```python
img_ori=Image.open('2.jpg')
np_img=np.array(img_ori)
img_gauss=np.zeros_like(np_img)
for i in range(3):
    img_gauss[:,:,i]=ndimage.gaussian_filter(np_img[:,:,i],sigma=10)
plt.imshow(np.hstack((img_ori,img_gauss)))
plt.title('')
plt.axis('off')
plt.show()
```

```python
img_bgr=cv2.imread('3.jpg')
img_rgb=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('')
plt.axis('off')
plt.show()
img_crop=img_rgb[300:1000,200:900]
plt.imshow(img_crop)
plt.title('')
plt.axis('off')
plt.show()
```

```python
np_img = np.array(Image.open('4.jpg'))
print('shape:', np_img.shape)
plt.figure(figsize=(6, 4))
plt.imshow(np_img)
plt.title('')
plt.axis('off')
plt.show()

scale = 0.5
img_resized = transform.resize(np_img, (int(np_img.shape[0]*scale), int(np_img.shape[1]*scale)), preserve_range=True).astype(np.uint8)
print('shape:', img_resized.shape)
plt.figure(figsize=(3, 2))
plt.imshow(img_resized)
plt.title('')
plt.axis('off')
plt.show()
```
